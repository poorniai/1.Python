```python
def AgeCategory():
    if(age<18):
        print("Children")
        cate="Children"
    elif(age<35):
        print("Adult")
        cate="Adult"
    elif(age<59):
        
        print("Citizen")
        cate="Citizen"
    else:
        print("Senior Citizen")
        cate="Senior Citizen"
    return cate
```

```python
age=int(input("Enter the age"))
agecate=AgeCategory()
```

```python
def OddEven():
    num=int(input("Enter Number"))
    if(num%2==0):
        print("Event Event")
        message="Event Number"
    elif(num%2==1):
        print("Odd Number")
        message="Odd number"
    return message
```

```python
OddEven()

 
```

```python
disMessage=OddEven()
```

```python
if(disMessage == 'Odd number'):
    print("The Given Number is Odd")
else:
    print("The Given Number is Even")
```
